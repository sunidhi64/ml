```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics   #see the performance, error kitna hai
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("stumarks.csv")
df
```

```python
m.scatter(df['Hours'], df['Scores'], color = 'k')
m.xlabel('Hours')
m.ylabel('Scores')
#We need to find a line such that error is reduced
```

```python
X = df.iloc[:, :-1].values #selects all col except last and all rows #features
Y = df.iloc[:, 1].values #selects all rows and last column #target value
```

```python

X
```

```python
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 20) #20% train and 80% test, 
#random state just tells how randomly data will be selected
#x_train cut for training from x, eise hi saare parameter
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train) #fitting the best line on the tests
```

```python
y_predict = regressor.predict(x_test)
```

```python
df_compare = pd.DataFrame({'Actual' : y_test, 'Predicted' : y_predict})
df_compare
```

```python
print('Mean Absolute error', metrics.mean_absolute_error(y_test, y_predict))
print('Mean Squared error', metrics.mean_squared_error(y_test, y_predict))
print('Root Mean squared error', n.sqrt(metrics.mean_squared_error(y_test, y_predict)))
```

```python
print(regressor.intercept_) #jaha line kaate
```

```python
print(regressor.coef_) #slope these two are parameters
```
